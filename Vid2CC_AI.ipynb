```python
#@title üé¨ Vid2CC-AI Control Panel
import os, uuid, ipywidgets as widgets
from IPython.display import clear_output, display
from google.colab import output

# --- Functionality ---
def run_process(b):
    video_path = video_input.value.strip()
    output_dir = output_input.value.strip()

    if not os.path.exists(video_path):
        with out:
            print(f"‚ùå Error: File not found at {video_path}")
            return

    # Build Command
    cmd = f'vid2cc-ai "{video_path}" --model {model_dropdown.value} -o "{output_dir}"'

    if sub_mode.value == "Soft Subtitles (--embed)":
        cmd += " --embed"
    elif sub_mode.value == "Hardcode (--hardcode)":
        cmd += " --hardcode"

    if translate_toggle.value:
        cmd += " --translate"

    with out:
        clear_output()
        print(f"üöÄ Running command: {cmd}\n")
        !{cmd}
        print("\nüéâ Task Finished!")

# --- UI Elements ---
style = {'description_width': 'initial'}
layout_full = widgets.Layout(width='95%')

video_input = widgets.Text(
    description='üìπ Video Path:',
    placeholder='/content/drive/MyDrive/video.mp4',
    style=style, layout=layout_full
)

output_input = widgets.Text(
    description='üìÅ Output Dir:',
    value='/content/drive/MyDrive/Vid2CC_Output',
    style=style, layout=layout_full
)

model_dropdown = widgets.Dropdown(
    options=['tiny', 'base', 'small', 'medium', 'large', 'turbo'],
    value='small',
    description='üß† Whisper Model:',
    style=style
)

sub_mode = widgets.Dropdown(
    options=['None (SRT only)', 'Soft Subtitles (--embed)', 'Hardcode (--hardcode)'],
    value='Soft Subtitles (--embed)',
    description='üìù Subtitle Style:',
    style=style
)

translate_toggle = widgets.Checkbox(
    value=False,
    description='Translate to English (--translate)',
    style=style
)

run_button = widgets.Button(
    description="Start Processing",
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''
    layout=widgets.Layout(width='95%', height='40px'),
    icon='play'
)

out = widgets.Output()

# --- Layout Assembly ---
run_button.on_click(run_process)

ui_box = widgets.VBox([
    widgets.HTML("<h2>Vid2CC-AI Dashboard</h2>"),
    video_input,
    output_input,
    widgets.HBox([model_dropdown, sub_mode]),
    translate_toggle,
    widgets.HTML("<br>"),
    run_button,
    widgets.HTML("<br>"),
    out
])

display(ui_box)
```
